```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

# Dataset for Classification with 2 clusters

```python
# Seed for reproducibility
np.random.seed(42)

# Number of points
n_points = 20

# Generate 50 points on one side of the hyperplane
X1 = np.random.normal(loc=[2, 2], scale=1, size=(10, 2))
y1 = np.zeros(10)  # Label for this class

# Generate 50 points on the other side of the hyperplane
X2 = np.random.normal(loc=[5, 5], scale=1, size=(10, 2))
y2 = np.ones(10)  # Label for this class

# Combine the points
X = np.vstack((X1, X2))
y = np.hstack((y1, y2))

# Shuffle the data
indices = np.arange(n_points)
np.random.shuffle(indices)
np.random.shuffle(indices)

X = X[indices]
y = y[indices]

# Create a DataFrame
data = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
data['Label'] = y

# Visualize the data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Binary Classification Data')
plt.show()

# Print the DataFrame head
print(data.head())
```

```python
filename = '/content/drive/MyDrive/Colab Notebooks/binary_classification_2d_data.xlsx'
data.to_excel(filename, index=False)
```

# Dataset for GMM with 3 clusters

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the first Gaussian distribution
mean1 = [2, 4]
cov1 = [[1, 0], [0, 1]]
num_samples1 = 300

# Parameters for the second Gaussian distribution
mean2 = [6, 8]
cov2 = [[1.0, 0], [0, 4.0]]
num_samples2 = 300

# Parameters for the third Gaussian distribution
mean3 = [12, 5]
cov3 = [[3.0, 0.6], [0.6, 1.0]]
num_samples3 = 300

# Generate samples
data1 = np.random.multivariate_normal(mean1, cov1, num_samples1)
data2 = np.random.multivariate_normal(mean2, cov2, num_samples2)
data3 = np.random.multivariate_normal(mean3, cov3, num_samples3)

# Plotting
plt.figure(figsize=(10, 6))

# Plot first Gaussian distribution
plt.scatter(data1[:, 0], data1[:, 1], color='blue', alpha=0.5, label='Distribution 1')

# Plot second Gaussian distribution
plt.scatter(data2[:, 0], data2[:, 1], color='red', alpha=0.5, label='Distribution 2')

# Plot third Gaussian distribution
plt.scatter(data3[:, 0], data3[:, 1], color='green', alpha=0.5, label='Distribution 3')

# Set equal scaling
plt.gca().set_aspect('equal', adjustable='box')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Two Gaussian Distributions')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# Create DataFrames for each distribution
df1 = pd.DataFrame(data1, columns=['X', 'Y'])
df1['label'] = 0

df2 = pd.DataFrame(data2, columns=['X', 'Y'])
df2['label'] = 1

df3 = pd.DataFrame(data3, columns=['X', 'Y'])
df3['label'] = 2

# Combine DataFrames
gaussian_df = pd.concat([df1, df2, df3], ignore_index=True)
gaussian_df
```

```python
filename = '/content/drive/MyDrive/Colab Notebooks/gaussian_dataset_3_class_data.xlsx'
gaussian_df.to_excel(filename, index=False)
```
