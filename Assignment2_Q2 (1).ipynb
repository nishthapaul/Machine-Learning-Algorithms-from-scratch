# First we will preprocess the data

1. We tried to find the percentage of missing data in the dataset, but there were no columns with null data.

```python
def missing_percentage(df):
    missing_data = pd.DataFrame(columns = ['column_name', 'percentage'])
    for col in df.columns:
        if df[col].isna().any():
            per = 100*df[col].isna().sum()/df.shape[0]
            df1 = pd.DataFrame({'column_name': col, 'percentage': per}, index = [0])
            missing_data = pd.concat([missing_data, df1])
    return missing_data
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

def normalize_dataframe(df):
    scaler = MinMaxScaler()

    normalized_array = scaler.fit_transform(df)

    normalized_df = pd.DataFrame(normalized_array, columns=df.columns)

    return normalized_df

normalized_house_data = normalize_dataframe(house_data)
```

# Applying Neural Networks using numpy library

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

class NeuralNetwork() :
    def __init__(self, learning_rate, no_of_iterations) :
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations
        self.weights = []
        self.bias = []
        self.layers = []

    def add_layer(self, no_of_neurons, activation_fn) :
        self.layers.append([no_of_neurons, activation_fn])

    def sigmoid(self, x) :
        return 1 / (1 + np.exp(-x))

    def leaky_relu(self, x, alpha=0.1) :
        return np.maximum(alpha * x, x)

    def tanh(self, x):
        return np.tanh(x)

    def leaky_relu_derivative(self, x, alpha=0.1) :
        return np.where(x > 0, 1, alpha)

    def sigmoid_derivative(self, x):
        sigmoid_x = self.sigmoid(x)
        return sigmoid_x * (1 - sigmoid_x)

    def tanh_derivative(self, x):
        return 1.0 - np.tanh(x) ** 2

    def print_structure(self, mat):
        for i, matrix in enumerate(mat):
            print(f"2D Vector {i + 1} Shape:", matrix.shape)

    def initializeWeights(self, no_of_features) :
        n = no_of_features
        for l in range(len(self.layers)) :
            m = self.layers[l][0]
            self.weights.append(np.random.normal(loc=0, scale=(1 / n), size=(n, m)))
            n = m

    def initializeBias(self, no_of_features) :
        for l in range(len(self.layers)) :
            self.bias.append(np.random.normal(loc=0, scale=(1 / self.layers[l][0]), size=(self.layers[l][0])))

    def normalise(self, vector) :
        norm = np.linalg.norm(vector)
        if norm == 0:
            return vector
        return vector / norm

    def add(self, list1, list2) :
        if not list1:
            return list2
        elif not list2:
            return list1
        result_array = np.array(list1) + np.array(list2)
        return result_array.tolist()

    def subtract_3d_lists(self, list1, list2):
        if not list1 and not list2:
            return []
        elif not list1:
            return [[-x for x in row] for row in list2]
        elif not list2:
            return list1

        if len(list1) != len(list2):
            raise ValueError("Number of 2D matrices must be the same for subtraction.")

        result_array = []
        for i, (m1, m2) in enumerate(zip(list1, list2)):
            if np.shape(m1) != np.shape(m2):
                if i == len(list1) - 1:
                    m2_2d = np.reshape(m2, (len(m2), 1))
                    result_array.append(np.array(m1) - m2_2d)
                else:
                    raise ValueError("2D matrices must have the same shape for subtraction.")
            else:
                result_array.append(np.array(m1) - np.array(m2))

        return result_array

    def divide(self, list1, divisor):
        result_array = np.array(list1) / divisor
        return result_array.tolist()

    def multiply(self, list1, a):
        result_array = np.array(list1) * a
        return result_array.tolist()

    def reverse(self, list1):
        return list1[::-1]

    def fit(self, X, y_actual) :

        no_of_features = len(X.columns)
        no_of_datapoints = len(X)

        # initialise weights and bias as gaussian distribution
        self.initializeWeights(no_of_features)
        self.initializeBias(no_of_features)

        for itr in range(self.no_of_iterations) :

            print("iterations ", itr)
            y_predicted = []

            gradient_cw_cummulative = []
            gradient_cb_cummulative = []

            for index, row in X.iterrows() :
                data_point = row.values

                a_of_all_layers = []
                z_of_all_layers = []

                # Forward propogation
                a = data_point
                for l in range(len(self.layers)) :
                    z = np.dot(self.weights[l].T, a) + self.bias[l]
                    if self.layers[l][1] == "leaky_relu" :
                        a = self.leaky_relu(z)
                    elif self.layers[l][1] == "sigmoid" :
                        a = self.sigmoid(z)
                    elif self.layers[l][1] == "tanh" :
                        a = self.tanh(z)
                    z_of_all_layers.append(z)
                    a_of_all_layers.append(a)

                y_predicted.append(a.item())

                # Backward propagation
                gradient_cw = []
                gradient_cb = []
                delta = []
                for l in range(len(self.layers) - 1, -1, -1) :
                    del_c_by_del_w = []

                    derivative_of_activation_fn = []
                    if self.layers[l][1] == "leaky_relu" :
                        derivative_of_activation_fn = self.leaky_relu_derivative(z_of_all_layers[l])
                    elif self.layers[l][1] == "sigmoid" :
                        derivative_of_activation_fn = self.sigmoid_derivative(z_of_all_layers[l])
                    elif self.layers[l][1] == "tanh" :
                        derivative_of_activation_fn = self.tanh_derivative(z_of_all_layers[l])

                    print("z_of_all_layers[l]")
                    print(z_of_all_layers[l])
                    print("derivative_of_activation_fn")
                    print(derivative_of_activation_fn)

                    if l == len(self.layers) - 1 :
                        delta = ( y_actual[index] - y_predicted[index] ) * derivative_of_activation_fn * -1
                        del_c_by_del_w = a_of_all_layers[l - 1] * delta
                    elif l == 0 :
                        delta = np.dot(self.weights[l + 1], delta) * derivative_of_activation_fn
                        del_c_by_del_w = np.outer(data_point, delta)
                    else :
                        delta = np.dot(self.weights[l + 1], delta) * derivative_of_activation_fn
                        del_c_by_del_w = np.outer(a_of_all_layers[l - 1], delta)

                    gradient_cw.append(del_c_by_del_w)

                    del_c_by_del_b = delta
                    gradient_cb.append(del_c_by_del_b)

                gradient_cw_cummulative = self.add(gradient_cw_cummulative, gradient_cw)
                gradient_cb_cummulative = self.add(gradient_cb_cummulative, gradient_cb)

            # Calculating cumulative gradient
            gradient_cw_cummulative = self.divide(gradient_cw_cummulative, no_of_datapoints)
            gradient_cb_cummulative = self.divide(gradient_cb_cummulative, no_of_datapoints)

            # update w and b via gradient descent
            print("prev weights and bias")
            print(self.weights)
            print("learning rate * gradient")
            print(self.multiply(self.reverse(gradient_cw_cummulative), self.learning_rate))
            self.weights = self.subtract_3d_lists(self.weights, self.multiply(self.reverse(gradient_cw_cummulative), self.learning_rate))
            self.bias = self.subtract_3d_lists(self.bias, self.multiply(self.reverse(gradient_cb_cummulative), self.learning_rate))
            print("Update weights and bias")
            print(self.weights)

            y_predicted = np.array(y_predicted)

            print(self.normalise(y_predicted))

            print(y_actual)


            # Calculate Loss
            loss = np.sum((y_actual - self.normalise(y_predicted)) ** 2) / (2 * no_of_datapoints)
            print("loss : " , loss)

    def predict(self, X) :
        y_predicted = []
        for index, row in X.iterrows() :
            data_point = row.values

            a = data_point
            for l in range(len(self.layers)) :
                z = np.dot(self.weights[l].T, a) + self.bias[l]
                if self.layers[l][1] == "leaky_relu" :
                    a = self.leaky_relu(z)
                elif self.layers[l][1] == "sigmoid" :
                    a = self.sigmoid(z)
                elif self.layers[l][1] == "tanh" :
                    a = self.tanh(z)

            y_predicted.append(a.item())

        return y_predicted
```

# Train the NN Model

```python
learning_rate = 0.01 # alpha
no_of_iterations = 15
model = NeuralNetwork(learning_rate, no_of_iterations)
model.add_layer(128, "sigmoid")
model.add_layer(1, "sigmoid")
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print(y_test)
print(y_pred)

y_test = model.normalise(y_test)
y_pred = model.normalise(y_pred)

print(y_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```
