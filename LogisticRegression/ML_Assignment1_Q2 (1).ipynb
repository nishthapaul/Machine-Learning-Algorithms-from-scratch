I created a LogisticRegression class with the following methods:
1. Sigmoid
2. fit (X_train, y_train)

Probability = sigmoid(w0 + w1x1 + w2x2 + w3x3 + ... + w9x9)
where w1 to w9 are the weights of the features, and x1 to x9 are the feature values for a particular player and w0 is bias.

We have to calculate Loss for every player and I used

cross entropy loss function ( L ) = 1/N ( summation i = 0 to N of ( yi * log pi + (1 - yi) * log (1 - pi) ) )

(N is number of players / train dataset)

Then we want to use gradient decent to iteratively find the optimal value of w0 and the 9 weights w.

w_new = w_old - learning_rate * dL / dw

bias_new = bias_old - learning_rate * dL / dbias

then I differentiated the loss function and derived

dL / dw = ( yi - sigmoid(w0 + w1x1 + ... + w9x9) ) * xi

dL / dbias = ( yi - sigmoid(w0 + w1x1 + ... + w9x9) )

3. predict (X_test)

This will predict the contribution_type of every test player by finding -

sigmoid(w0 + w1x1 + ... + w9x9)

If this values is less than 0.5, then it belongs to category 0, otherwise 1.

```python
import numpy as np
class LogisticRegression() :
  def __init__(self, learning_rate, no_of_iterations) :
    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations
    self.weight = None
    self.bias = None

  def sigmoid(self, x) :
    return 1 / (1 + np.exp(-x))

  def fit(self, X, y) :
    no_of_features = len(X.columns)
    N = len(X)
    self.weight = np.zeros(no_of_features)
    self.bias = 0

    convergence_threshold = 0.0001

    for itr in range(self.no_of_iterations) :
      # differentiate Loss function wrt w (feature vector)
      dL_dw = np.zeros(no_of_features)
      for index, row in X.iterrows():
        t = self.bias + np.dot(self.weight, row.values)
        dL_dwi = (y.loc[index] - self.sigmoid(t)) * row.values
        np.add(dL_dw, dL_dwi, out=dL_dw)
      dL_dw = dL_dw / N

      # differentiate Loss function wrt w0 (bias)
      dL_dbias = 0
      for index, row in X.iterrows():
        t = self.bias + np.dot(self.weight, row.values)
        dL_dbiasi = y.loc[index] - self.sigmoid(t)
        dL_dbias = dL_dbias + dL_dbiasi
      dL_dbias = dL_dbias / N

      new_bias = self.bias - self.learning_rate * dL_dbias
      new_weight = np.subtract(self.weight, self.learning_rate * dL_dw)

      weight_change = np.linalg.norm(new_weight - self.weight)
      if weight_change < convergence_threshold:
        print(f"Converged after {itr} iterations.")
        break

      self.bias = new_bias
      self.weight = new_weight

  def predict(self, X) :
    y_pred = pd.Series()
    for index, row in X.iterrows():
      t = self.bias + np.dot(self.weight, row.values)
      if self.sigmoid(t) < 0.5 :
        y_pred = pd.concat([y_pred, pd.Series([0])], ignore_index=True)
      else :
        y_pred = pd.concat([y_pred, pd.Series([1])], ignore_index=True)
    return y_pred
```

```python
learning_rate = 0.01 # alpha
no_of_iterations = 100
model = LogisticRegression(learning_rate, no_of_iterations)
model.fit(X_train, y_train)
```
